#### Split Task

 first task is to split each row of data.csv into text file , 
the file name is the id column and the text is the content of text column

here is the datset , want the name of outp of each text is the id number 
and the content is the text inside each text for the same id

```python
# Load EDA Pkgs
import pandas as pd 
import numpy as np
```

```python
# Load Dataset
df = pd.read_csv("last.csv")
```

```python
df.head()
```

```python
# Shape
df.shape
```

```python
df.columns
```

```python
df = df[['id', 'text']]
```

```python
df.head()
```

#### Method  1
+ Step 1 Group all By ID
+ Step 2  Join All Text of the Same ID As One


```python

grouped_df =  df.groupby('id').agg({
    'text': lambda x: ', '.join(x),
})
```

```python
# Shape
grouped_df.shape
```

```python
# Len of Unique Rows
len(df['id'].unique().tolist())
```

```python
# Len of Unique Rows For Index
len(grouped_df.index.tolist())
```

#### Narrative
+ Both have the same len

```python
# Already Sorted
grouped_df.index
```

```python
id_list = grouped_df.index.tolist()
```

```python
id_list
```

```python
for i,row in zip(id_list,grouped_df.values):
    file_name = '{}.txt'.format(i)
    row.tofile(file_name, sep=",", format="%s")
    i+=1
    
```
